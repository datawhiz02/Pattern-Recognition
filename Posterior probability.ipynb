```python
#STEP 3
testing_x["posterior_probability1_" + str(subset1_y.unique()[0])] = 0.0

#calculating prior and likelihood for 1st class

for index, row in testing_x.iterrows():
    for value in subset1_y.unique():
        
#Considered testing with 2 columns because cell takes more than 20 seconds to run otherwise.Using subset of features.

        prior_probability = sum(subset1_y == value) / len(subset1_y)
        predictor_withvalue = subset1_x[subset1_y == value]
        for col in subset1_x.columns[:2]:
            total_likelihood = 1
                        
            likelihood = (sum(predictor_withvalue[col] == row[col]) / sum(subset1_y == value))
            total_likelihood *= likelihood
             
        posterior_probability = prior_probability * total_likelihood
               
        testing_x.loc[testing_x.index == index ,"posterior_probability1_" + str(value)] = posterior_probability
```
